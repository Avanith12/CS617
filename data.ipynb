```python
import pandas as pd
import numpy as np
```

```python
students_path = r"C:\Users\kanam\OneDrive\Desktop\Projects\cs617\Students.csv"
faculty_path  = r"C:\Users\kanam\OneDrive\Desktop\Projects\cs617\Faculty.csv"
```

```python
students = pd.read_csv(students_path, skiprows=3)
faculty  = pd.read_csv(faculty_path,  skiprows=2)
```

```python
students.head(), faculty.head()
```

```python
sem_students = ["F16","Sp17","F17","Sp18","F18","Sp19","F19","Sp20","F20","Sp21","F21","Sp22","F22","Sp23","F23"]
sem_faculty  = ["F16","Sp17","F17","Sp18","F18","Sp19","F19","Sp20","F20","Sp21","F21","Sp22","F22","Sp23"]
```

```python
students_total = students[sem_students].sum(numeric_only=True)
```

```python
students_total
```

```python
faculty_total = faculty[sem_faculty].sum(numeric_only=True)
```

```python
faculty_total
```

```python
common = students_total.index.intersection(faculty_total.index)
```

```python
students_s = students_total.loc[common]
faculty_s = faculty_total.loc[common]
```

```python
students_idx = (students_s / students_s.iloc[0]) * 100
faculty_idx = (faculty_s / faculty_s.iloc[0]) * 100
```
